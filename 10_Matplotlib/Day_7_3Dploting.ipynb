````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
````

````python
# Generate evenly spaced numbers
x = np.linspace(1, 5, 100)
y = np.linspace(6, 10, 100)

# Create a meshgrid
X, Y = np.meshgrid(x, y)

# Sample function
Z = X * Y

# Create a filled contour plot
plt.contourf(X, Y, Z, cmap='viridis')  # Corrected colormap
plt.colorbar(label='Z value')  # Add a colorbar with a label
plt.xlabel('X')  # Label for the X-axis
plt.ylabel('Y')  # Label for the Y-axis
plt.title('Contour Plot')  # Title of the plot
plt.show()  # Display the plot
````

### Contour Plot (contourf)
- A contour plot is a graphical representation used to show 3D data in two dimensions, using contour lines or filled regions. It is typically used to represent elevation maps, temperature distributions, and other scalar fields.

- contour(): Draws contour lines.
- contourf(): Draws filled contours (like a heatmap).

### ** What is a Residual Plot?**  
A **residual** is the difference between the actual (`y`) value and the predicted (``) value from a regression model:  
\[
Residual = y - \hat{y}
\]
A good regression model will have residuals randomly scattered around zero.

---

### ** Example: Residual Plot using `sns.residplot()`**
```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y = 2*x + np.random.normal(0, 2, size=len(x))  # Linear data with noise

# Create residual plot
sns.residplot(x=x, y=y, lowess=True, color="blue")

# Add labels
plt.title("Residual Plot")
plt.xlabel("X values")
plt.ylabel("Residuals")
plt.axhline(0, color='red', linestyle='--')  # Zero residual reference line
plt.show()
```

---

### ** How to Interpret a Residual Plot?**
1. **Randomly Scattered Residuals ( Good Fit)**  
   - If residuals are randomly scattered around zero, the regression model is a **good fit**.  

2. **Curved Pattern in Residuals ( Bad Fit)**  
   - If there's a pattern (e.g., U-shape), it suggests that a **linear model is not appropriate**, and you might need a polynomial or another regression model.

---

### ** Use Cases of Residual Plots**
- Checking if a **linear regression model is appropriate** for the data.
- Detecting **heteroscedasticity** (unequal variance of residuals).
- Identifying **outliers** in regression analysis.

````python
# Sample DataFrame
np.random.seed(42)
df = pd.DataFrame({
    'StudentID': range(1, 101),
    'ExamScore': np.random.randint(50, 100, size=100)  # Random scores between 50 and 100
})

# Create Performance Category
df['Performances'] = ['Good' if score > 75 else 'Average' for score in df['ExamScore']]

# Create FacetGrid with histograms of ExamScore based on Performances
g = sns.FacetGrid(df, col="Performances", height=5, aspect=1.2)
g.map(plt.hist, "ExamScore", bins=10, color="skyblue", edgecolor="black")

plt.show()
````

### ** When to Use What?**
-  Use Frequency Histogram when you want to see the exact counts of data points.
-  Use Density Histogram when comparing datasets of different sizes or when visualizing a probability distribution.

````python
# Generate sample data
np.random.seed(42)
x = np.random.normal(50, 10, 500)  # Random values centered at 50
y = np.random.normal(60, 15, 500)  # Random values centered at 60

# Create scatter plot with KDE contours
sns.kdeplot(x=x, y=y, fill=True, cmap="Blues", thresh=0.05)

# Add scatter plot for individual points
plt.scatter(x, y, color='red', alpha=0.3, s=10)

# Show plot
plt.title("KDE Scatter Plot")
plt.xlabel("X-axis Values")
plt.ylabel("Y-axis Values")
plt.show()
````
